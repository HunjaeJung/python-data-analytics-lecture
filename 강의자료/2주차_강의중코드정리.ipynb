```python
# -*- coding: utf-8 -*-

# 저번 주 도전과제 해답 같이 풀어보기
coffee_menu = [
    {
        "name": "아메리카노",
        "price": 4000
    },
    {
        "name": "카페라떼",
        "price": 5000
    },
    {
        "name": "카페모카",
        "price": 6000
    }
]

# 2중 for문을 사용한 해답
def calculate_total_price(coffees):

    result = 0

    for menu in coffee_menu:
        for target in coffees:
            if menu['name'] == target:
                result = result + menu['price']

    return result

print calculate_total_price(["카페모카", "카페라떼"])

# if .. in 을 이용한 풀의법
# 2중 for 문을 간략화 할 수 있습니다.
def calculate_total_price(coffees):
    result = 0
    for menu in coffee_menu:
        if menu['name'] in coffees:
            result = result + menu['price']

    return result

print calculate_total_price(["카페모카", "카페라떼"])


# if .. in 에 대한 간략한 설명
fruits = ["사과", "포도", "수박"]
if "사과" in fruits:
    print "있습니다."
else:
    print "없습니다."

################################################################
# 함수는 딕셔너리나 리스트 같은 다양한 형태로 return이 가능합니다.
def sum_func(a, b):
#     res = {
#         "a": a,
#         "b": b
#     }
#     return res
    return a + b

# return을 이용하여 sum_func() 함수를 중첩해서 사용한 모습
print sum_func(100, sum_func(sum_func(1, 10), 20))


# 함수도 변수로 취급할 수 있습니다.
def sum_func(a, b):
    return a + b

def minus_func(a, b):
    return a - b

print sum_func  # 함수가 가르키는 메모리 주소값을 보여줍니다.


# type() 내장함수를 이용하면 변수의 자료형(type)을 확인할 수 있습니다.
a = "hi"
print type(a)
print type(sum_func)

# 함수도 하나의 변수(자료형)이므로, 함수의 매개변수에 넣을 수 있습니다.
def calc(num1, num2, what_to_do):

    return what_to_do(num1, num2)

print calc(10, 20, sum_func)
print calc(10, 20, minus_func)


# 또한 return 할 수도 있습니다.
# 더불어 함수 내에서 함수를 선언하는 것도 가능합니다.
def get_calc_module(module_name):

    def inner_sum_func(a, b):
        return a + b
    def inner_minus_func(a, b):
        return a - b

    if module_name == "sum":
        return inner_sum_func
    elif module_name == "minus":
        return inner_minus_func

result_fucntion = get_calc_module("sum")
print result_fucntion
print result_fucntion(100, 200)


# for 문을 이용해 factorial을 구현한 것입니다.
# 이를 재귀 호출(recursive)를 이용해 구현할 수도 있을 것입니다.
def factorial_for(n):
    result = 1
    for i in range(1, n+1):
        result = result * i

    return result

print factorial_for(5)


# 피보나치 수열 구하기를 재귀 호출을 통해 구현한 모습입니다.
def fib(n):

    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fib(n-1) + fib(n-2)

################################################################
# 가변길이 매개변수(*args)의 모습입니다.
# 아래는 매개변수 이름을 *nums로 지정하였지만, 관습상 *args라고 명명하는 것이 일반적입니다.
def var_sum_func(*nums):

    result = 0
    for n in nums:
        result = result + n

    return result

print var_sum_func(1, 2, 3, 4, 5, 2)


# 키워드 가변길이 매개변수의 예제입니다. 관습상 **kwargs라고 명명합니다.
def fruits(**kwargs):

    # print kwargs
    for key, val in kwargs.items():  # 딕셔너리를 for 문으로 순회할 때에는 .items() 라는 내장 변수를 사용합니다.
        print key, val

fruits(grape=3000, lemon=5000, apple=1000, watermelon=10000)


################################################################
# List comprehension 입니다.

# 입력된 가격 리스트의 2배에 해당하는 리스트를 만들어봅시다.
new_double_price = []
prices = [1000, 2000, 3000, 4000, 5000]

# 기본적으로 for 문으로는 이렇게 구현됩니다.
for price in prices:
    new_double_price.append(price * 2)

print new_double_price
# print sum(prices)  # sum() 내장함수를 통해 리스트의 모든 값을 더할 수 있습니다.

# list comprehension을 이용하면 이러한 형태로 만들어집니다.
double_price = [price * 2 for price in prices]
print double_price


# 저번주 도전과제였던 커피 총 가격 합 구하기를 이렇게 list comprehension으로 구현할 수도 있습니다.
print sum([menu['price'] if menu['name'] in ["아메리카노", "카페모카"] else 0 for menu in coffee_menu])


# 주어진 리스트 중, 홀수는 두배로, 짝수는 그대로 리스트에 복사하는 예제입니다.
sample = [1, 3, 5, 7, 9, 3, 4, 5]
# 예상 결과: [2, 6, 10, 14, 18, 6, 4, 10]

result = [n if n % 2 == 0 else n * 2 for n in sample]
print result

################################################################
# 참(True)와 거짓(False), 논리 연산에 대한 설명입니다.

a = True
print not not not a  # not으로 뒤집을 수 있습니다.

a = 1
b = 2

# and 연산과 or 연산이 존재합니다.
if a == 1 and b == 3:
    print "둘 다 맞네요.."


# 논리연산을 이용한 '여행을 갈 수 있는지' 알아보는 함수 예제입니다.
# is_weekday: 주중인지 아닌지
# is_vacation: 휴가 기간인지 아닌지
def can_travel(is_weekday, is_vacation):

    if not is_weekday or is_vacation:
        return True
    else:
        return False

# 연산자 우선 순위에 대한 예제입니다.
a = 1 + 3
if a == 4 and a >= 1*2:
    print "가장 먼저 우항의 곱셈 연산이 이루어지고, 좌항의 평등 연산이 이루어진 이후에 가운데 논리 연산이 이루어졌습니다."


################################################################
# 문자열과 리스트는 형제입니다.

pizza_list = ["P", "a", "p", "a"] # ...
print pizza_list[-1]  # 이렇게 음수로 접근할 수도 있고,
print pizza_list[1:3]  # :를 이용해 구간을 정할 수도 있습니다.
pizza = "Papa jones"

print pizza[0:-6]

print len(pizza)  # 내장함수 len()을 이용하면 리스트나 문자열의 길이를 알 수 있습니다.
for ch in pizza:
    print ch  # 리스트와 문자열은 형제기에, 이렇게 for..in 으로 처리할 수도 있습니다.


# 문자열 중, 'e'가 몇 개 들어있는제 카운트하는 에제입니다. 
def count_alphabet(input):
    count = 0

    for ch in input:
        if ch == 'e':
            count += 1

    return count

print count_alphabet("eeeeeeeeee")


# 문자열을 무조건 두 번 씩 쓰게 해주는 함수 예제입니다.
def double_alphabet(input):
    result = ""

    for ch in input:
        result += ch * 2

    return result

print double_alphabet("hello!!")


# 문자열 formatting 예제입니다. 
count = 3
print "나는 " + str(count) + "개의 사과를 먹었습니다."  # 기존에는 + 연산을 이용해 문자열을 합쳤을 것입니다. str()을 사용하면 숫자를 문자열로 캐스팅(변환)할 수 있습니다.

# 문자열 formatting시에는 *args와 **kwargs 모두 사용 가능합니다.
apple_mania = "나는 {cnt}개의 사과를 먹었습니다."
print apple_mania.format(cnt=100)

fruits_mania = "나는 {0}개의 사과를 먹었고, {1}개의 포도를 먹었습니다. 그리고 {lemon}개의 레몬을 먹을 것입니다."
print fruits_mania.format(3, 5, lemon=10)

# 문자열 formatting과 for문을 활용한 예제입니다.
def pigs(n):
    for i in range(1, n+1):
        print "{0}마리의 돼지가 {1}하고 웁니다.".format(i, "꿀"*i)

pigs(10)


# 각종 문자열 내장 함수들에 대한 활용 예제입니다.
a = "hello"
dash = "-"
print a.lower().upper()  # ->소문자->대문자

print a.count("l")  # 갯수 세기
print a.find('z')  # 찾기 (찾으면 해당 위치, 없으면 -1 리턴)

print dash.join(a)  # 삽입(합치기)

print "      Hello!!    ".strip()  # 양쪽 공백 제거

print "sugar is so sweet".replace("sugar", "salt").replace("sweet", "salty")  # 문자열 치환(replace)

# split()과 join()을 동시에 활용한 예제.
gump = "Life is like a box of chocolate"
splited_gump = gump.split()
print splited_gump
original_gump = "/".join(splited_gump)
print original_gump
```
